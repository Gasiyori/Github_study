## Zigbee 프로토콜 패킷 분리

```python
import pandas as pd

log_zgb = pd.read_csv('RSSI_20_1000.log', delimiter=',', names=["Time","Id","Type","Data", "Start delimiter", "Length", "Frame type",
                                                            "64-bit source address", "16-bit source address", "Receive options",
                                                            "RF data", "Checksum"])
```

```python
log_zgb = log_zgb.drop(index = 0, axis = 0)
```

```python
log_zgb.set_index("Id", inplace=True) #id로 인덱스 재지정
```

```python
log_zgb.Data #패킷 데이터 확인
```

```python
log_zgb["Start delimiter"] = log_zgb.Data.str[:2]
log_zgb["Length"] = log_zgb.Data.str[2:6]
log_zgb["Frame type"] = log_zgb.Data.str[6:8]
log_zgb["64-bit source address"] = log_zgb.Data.str[8:24]
log_zgb["16-bit source address"] = log_zgb.Data.str[24:28]
log_zgb["Receive options"] = log_zgb.Data.str[28:30]
log_zgb["RF data"] = log_zgb.Data.str[30:-2] #페이로드 사이즈에 따라 위치 재조정 필요. 이 경우 20byte
log_zgb["Checksum"] = log_zgb.Data.str[-2:] #각각의 사이즈에 맞게 스플릿
```

```python
log_zgb = log_zgb.drop(columns = "Data") #헷갈리니까 삭제
```
