# 누락 및 비정상 패킷 비율

재전송해야하는 패킷을 구하기 위해서는 정상 패킷 수신 비율을 알아야 됨.

누락 패킷 = 전체 패킷 - 수신 패킷

이 경우, 전체 패킷은 1000개 이므로 수신 패킷의 정상 유무와 상관없이 1000에서 수신 패킷을 빼면 됨

shape[0]은 행 개수, shape[1]은 열 개수이므로 shape[0]을 이용해 쉽게 구할 수 있음

```python
whole_packet = 1000

missed_packet = whole_packet - log_20_4_5.shape[0]

missed_packet
```

누락 및 비정상 패킷 수신 비율 = (누락 패킷 수 + 비정상 패킷 수) / 전체 패킷 수

RF data(=payload)의 값이 변하면 Length의 길이가 바뀌므로 비정상 패킷 판단 가능

```python
abnormal_packet_rate = (missed_packet + log_20_4_5[log_20_4_5["Length"] != 25].shape[0]) / whole_packet

abnormal_packet_rate
```

전체 데이터 1,000,000 바이트를 보낸다고 가정했을 때,

20바이트씩 끊어서 보낸 경우 50,000번의 전송을 하고 (`1,000,000 / 20 = 50,000`) (전체 데이터 / 바이트 수)

0.008의 확률로 재전송을 해야하므로 재전송 횟수는 400회 (`(1,000,000 / 20) * 0.008 = 400`) ( (전체 데이터 / 바이트 수) * 이상 패킷 비율)

따라서 총 전송 횟수는 50,400회 (`(1,000,000 / 20) + (1,000,000 / 20) * 0.008`) (전체 데이터 / 바이트 수) + ((전체 데이터 / 바이트 수) * 이상패킷 비율)

총 전송 횟수 = (1,000,000 / 분할 바이트 수) + ((1,000,000 / 분할 바이트 수 * 누락 및 비정상 패킷 비율))

## 반복문으로 데이터프레임 호출해보기

문자열 내부에 .이 들어가 있으면 오류가 생긴다.

예를 들자면, `log_20byte_distance_0m`는 괜찮은데 `log_20byte_distance_4.5m`는 오류가 생김

따라서 cm로 수정해야 하거나 언더바(_)로 대체해야 할듯

> 언더바로 대체됨

추가) 70byte의 27m와 같이 내부 데이터가 전혀 없는 경우 오류 발생

이러한 데이터를 처리하기 위한 방법이 필요함

> if not으로 비어있는지 여부 체크해서 해결

추가2) 한번에 처리하는게 좀 어려움

eval exec 함수로 처리 등등 다 해봤는데 잘 안됨.

> globals로 개별적으로 처리해서 동시에 처리 끝냄<br>
문자열이 포함된 변수 이름의 경우 이렇게 접근하는 편이 빠름

```python
import sys
mod = sys.modules[__name__]

# eval('log_' + str(i) + 'byte_distance_' + str(j))

for i in packet_byte:
    for j in packet_distance_name:
        if not globals()['log_' + str(i) + 'byte_distance_' + str(j)].empty: #비어있는 데이터가 아니면 적절히 처리
            globals()['log_' + str(i) + 'byte_distance_' + str(j)] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].drop(index = 0, axis = 0)
            globals()['log_' + str(i) + 'byte_distance_' + str(j)].set_index("Id", inplace=True)
            
            #패킷 헤더 규격에 맞게 분리
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Start delimiter"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[:2]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Length"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[2:6]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Frame type"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[6:8]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["16-bit source address"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[8:12]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["RSSI"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[12:14]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Options"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[14:16]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["RF data"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[16:-2]
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Checksum"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].Data.str[-2:]
            
            #필요없는 기존 데이터 처리
            globals()['log_' + str(i) + 'byte_distance_' + str(j)] = globals()['log_' + str(i) + 'byte_distance_' + str(j)].drop(columns = "Data")
            
            #Length 16진수>>10진수 변환
            globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Length"] = globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Length"].apply(func)
```

전체 데이터 1,000,000 바이트를 보낸다고 가정했을 때,

20바이트씩 끊어서 보낸 경우 50,000번의 전송을 하고 (`1,000,000 / 20 = 50,000`) (전체 데이터 / 바이트 수)

0.008의 확률로 재전송을 해야하므로 재전송 횟수는 400회 (`(1,000,000 / 20) * 0.008 = 400`) ( (전체 데이터 / 바이트 수) * 이상 패킷 비율)

따라서 총 전송 횟수는 50,400회 (`(1,000,000 / 20) + (1,000,000 / 20) * 0.008`) (전체 데이터 / 바이트 수) + ((전체 데이터 / 바이트 수) * 이상패킷 비율)

총 전송 횟수 = (1,000,000 / 분할 바이트 수) + ((1,000,000 / 분할 바이트 수 * 누락 및 비정상 패킷 비율))

**이후의 모든 데이터는 1,000,000바이트로 가정함**

```python
import math

for i, k in zip(packet_byte, packet_byte_list):
    for j, m in zip(packet_distance_name, packet_distance): #거리 목록, 분할 바이트 수
        if not globals()['log_' + str(i) + 'byte_distance_' + str(j)].empty: 
#             print(f"{i}, {j}") #디버깅용
            send_data = 1000000
            whole_packet = 1000
            
            missed_packet = whole_packet - globals()['log_' + str(i) + 'byte_distance_' + str(j)].shape[0] #누락 패킷 탐지
#             print(missed_packet) #누락 패킷 개수
    
            abnormal_packet_rate = (missed_packet + len(globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Length"][globals()['log_' + str(i) + 'byte_distance_' + str(j)]["Length"] != int(i) + 5])) / whole_packet
            #비정상 패킷 비율 = (누락 패킷 + 길이가 다른 수신 패킷) / 전체 패킷
                        
            total_send = math.ceil((send_data / int(i)) + (send_data / int(i)) * abnormal_packet_rate)
#             print(total_send)
            #재전송을 포함한 전체 전송율 (소수로 나눠지는 경우 올림처리)
            
            packet_table[m][k] = total_send
#             테이블에 집어넣음
```
